# 5. Tasks to Be Performed:

1. Create a Git Flow workflow architecture on Git

2. Create all the required branches

3. starting from e feature branch, push the branch to the master, following the architecture

4. Push an urgent.txt on master using hotfix


```bash
# 1. Create local repo
mkdir git-flow-project
cd git-flow-project
git init

# 2. Create Git Flow README
echo "# Git Flow Workflow

## Branches:
- master: Production-ready code
- develop: Integration branch for features
- feature/*: Feature branches from develop
- release/*: Release branches from develop
- hotfix/*: Emergency fixes from master

## Workflow:
1. Feature branches → develop
2. Develop → release
3. Release → master + develop
4. Hotfix → master + develop
" > README.md

git add README.md
git commit -m "Add Git Flow workflow architecture in README.md"

# 3. Create and switch between all required branches
git checkout -b develop
git checkout -b feature/login-system
git checkout develop
git checkout -b release/v1.0
git checkout master
git checkout -b hotfix/urgent-fix

# 4. Add a feature in feature/login-system
git checkout feature/login-system
echo "<!-- Login Feature HTML -->" > login.html
git add login.html
git commit -m "Add login feature in feature/login-system"

# 5. Merge feature into develop
git checkout develop
git merge feature/login-system -m "Merge login feature into develop"

# 6. Merge develop into release
git checkout release/v1.0
git merge develop -m "Prepare release v1.0"

# 7. Merge release into master (production release)
git checkout master
git merge release/v1.0 -m "Release v1.0 to production"

# 8. Merge release back into develop
git checkout develop
git merge release/v1.0 -m "Merge release back into develop"

# 9. Hotfix: add urgent.txt and apply to master
git checkout hotfix/urgent-fix
echo "URGENT HOTFIX - SECURITY PATCH" > urgent.txt
git add urgent.txt
git commit -m "Add urgent hotfix to urgent.txt"

# 10. Merge hotfix into master
git checkout master
git merge hotfix/urgent-fix -m "Apply urgent hotfix to master"

# 11. Merge hotfix into develop
git checkout develop
git merge hotfix/urgent-fix -m "Apply urgent hotfix to develop"

# 12. Set up GitHub remote repo
git remote add origin git@github.com:viku01999/git-flow-project.git

# 13. Push all branches to GitHub
git push -u origin master
git push -u origin develop
git push -u origin feature/login-system
git push -u origin release/v1.0
git push -u origin hotfix/urgent-fix

```
