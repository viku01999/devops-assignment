# Theory Questions

## 1. What is the difference between a function and a method in Python?
Answer: Function is defined using `def` and works independently. Method is a function inside a class and works with objects.
Example:

```python
def func():
    print("Function")

class A:
    def method(self):
        print("Method")
```

## 2. Explain the concept of function arguments and parameters in Python.
Answer: Parameters are variables in function definition, arguments are values passed.
Example:

```python
def add(a, b):
    return a + b

print(add(2,3))
```

## 3. What are the different ways to define and call a function in Python?
Answer: Functions can be defined using `def` and `lambda`.
Example:

```python
def f1():
    print("Hello")

f1()

f2 = lambda x: x+1
print(f2(5))
```

## 4. What is the purpose of the `return` statement in a Python function?
Answer: It sends value back from function.
Example:

```python
def add(a,b):
    return a+b
```

## 5. What are iterators in Python and how do they differ from iterables?
Answer: Iterable is object that can be looped. Iterator gives next value using `next()`.
Example:

```python
lst = [1,2,3]
it = iter(lst)
print(next(it))
```

## 6. Explain the concept of generators in Python and how they are defined.
Answer: Generators produce values one by one using `yield`.
Example:

```python
def gen():
    yield 1
    yield 2
```

## 7. What are the advantages of using generators over regular functions?
Answer: They save memory and generate values one by one.
Example:

```python
def gen():
    yield 1
```

## 8. What is a lambda function in Python and when is it typically used?
Answer: Lambda is small anonymous function used for short operations.
Example:

```python
f = lambda x: x*2
```

## 9. Explain the purpose and usage of the `map()` function in Python.
Answer: `map()` applies a function to all elements.
Example:

```python
print(list(map(lambda x:x*2,[1,2,3])))
```

## 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
Answer:
map → transforms data
filter → selects data
reduce → combines data
Example:

```python
from functools import reduce
print(list(map(lambda x:x*2,[1,2])))
print(list(filter(lambda x:x>1,[1,2])))
print(reduce(lambda x,y:x+y,[1,2,3]))
```

# Practical Questions

## 1. Sum of even numbers**

```python
def even_sum(lst):
    return sum(x for x in lst if x%2==0)

print(even_sum([1,2,3,4,5]))
```

## 2. Reverse string

```python
def rev(s):
    return s[::-1]

print(rev("hello"))
```

## 3. Squares of list

```python
def square(lst):
    return [x*x for x in lst]

print(square([1,2,3]))
```

## 4. Prime check (1 to 200)

```python
def is_prime(n):
    if n<2:
        return False
    for i in range(2,n):
        if n%i==0:
            return False
    return True

for i in range(1,201):
    if is_prime(i):
        print(i)
```

## 5. Fibonacci iterator class

```python
class Fib:
    def __init__(self,n):
        self.n=n
        self.a=0
        self.b=1
        self.count=0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count>=self.n:
            raise StopIteration
        val=self.a
        self.a,self.b=self.b,self.a+self.b
        self.count+=1
        return val

for i in Fib(5):
    print(i)
```

## 6. Generator powers of 2

```python
def power(n):
    for i in range(n+1):
        yield 2**i

for i in power(5):
    print(i)
```

## 7. Generator read file

```python
def read_file(file):
    with open(file) as f:
        for line in f:
            yield line

for line in read_file("file.txt"):
    print(line)
```

## 8. Lambda sort tuples

```python
lst = [(1,3),(2,1),(4,2)]
lst.sort(key=lambda x:x[1])
print(lst)
```

## 9. Celsius to Fahrenheit using map

```python
c = [0,10,20]
f = list(map(lambda x:(x*9/5)+32,c))
print(f)
```

## 10. Remove vowels using filter

```python
s = "hello world"
res = "".join(filter(lambda x:x not in "aeiou",s))
print(res)
```

## 11. Book shop problem
