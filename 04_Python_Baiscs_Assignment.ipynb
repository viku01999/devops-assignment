```py
a = [1,2,3]
a[0] = 10

```

- Immutable objects cannot be changed after creation.

```py
t = (1,2,3)
# t[0] = 10 will give error
```

# Task6: Describe the different types of loops in Python and their use cases with examples

- for loop → used when number of iterations is known
```py
for i in range(5): print(i)
```

- while loop → used when condition is checked

```py
x = 0
while x < 5:
  print(x)
  x += 1
```
